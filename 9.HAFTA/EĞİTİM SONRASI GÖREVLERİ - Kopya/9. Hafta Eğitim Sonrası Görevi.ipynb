# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import random
import time

class Kumanda:
    def __init__(self, tv_durum="Kapalı", tv_ses=10, kanal_listesi=["TRT"], aktif_kanal="TRT"):
        self.tv_durum = tv_durum
        self.tv_ses = tv_ses
        self.kanal_listesi = kanal_listesi
        self.aktif_kanal = aktif_kanal
    
    def tv_ac(self):
        if self.tv_durum == "Açık":
            print("Televizyon zaten açık.")
        else:
            print("Televizyon açılıyor...")
            self.tv_durum = "Açık"
    
    def tv_kapat(self):
        if self.tv_durum == "Kapalı":
            print("Televizyon zaten kapalı.")
        else:
            print("Televizyon kapanıyor...")
            self.tv_durum = "Kapalı"
    
    def ses_ayarlari(self):
        while True:
            cevap = input("Sesi artır: '+'\nSesi azalt: '-'\nÇıkış: 'q'\nSeçiminiz: ")
            
            if cevap == "+":
                if self.tv_ses < 100:
                    self.tv_ses += 1
                    print(f"Ses: {self.tv_ses}")
                else:
                    print("Ses zaten maksimum seviyede.")
            elif cevap == "-":
                if self.tv_ses > 0:
                    self.tv_ses -= 1
                    print(f"Ses: {self.tv_ses}")
                else:
                    print("Ses zaten minimum seviyede.")
            elif cevap == "q":
                print("Ses ayarlarından çıkılıyor...")
                break
            else:
                print("Geçersiz giriş, lütfen tekrar deneyin.")
    
    def kanal_ekle(self, kanal_ismi):
        print("Kanal ekleniyor...")
        time.sleep(1)
        self.kanal_listesi.append(kanal_ismi)
        print(f"{kanal_ismi} kanalı eklendi.")
    
    def kanal_sayisini_ogrenme(self):
        return len(self.kanal_listesi)
    
    def rastgele_kanal(self):
        rastgele = random.choice(self.kanal_listesi)
        self.aktif_kanal = rastgele
        print(f"Rastgele kanala geçildi: {self.aktif_kanal}")
    
    def televizyon_bilgileri(self):
        print(f"TV Durum: {self.tv_durum}\nSes: {self.tv_ses}\nKanal Listesi: {self.kanal_listesi}\nAktif Kanal: {self.aktif_kanal}")
kumanda = Kumanda()

while True:
    print("\nKumanda Uygulaması")
    print("1. Tv Aç")
    print("2. Tv Kapat")
    print("3. Ses Ayarları")
    print("4. Kanal Ekle")
    print("5. Kanal Sayısını Öğrenme")
    print("6. Rastgele Kanala Geçme")
    print("7. Televizyon Bilgileri")
    print("Çıkmak için 'q' tuşlayın")

    islem = input("Hangi işlemi yapmak istiyorsunuz?: ")

    if islem == "q":
        print("Programdan çıkılıyor...")
        break
    elif islem == "1":
        kumanda.tv_ac()
    elif islem == "2":
        kumanda.tv_kapat()
    elif islem == "3":
        kumanda.ses_ayarlari()
    elif islem == "4":
        kanal_ismi = input("Eklemek istediğiniz kanalın ismi nedir?: ")
        kumanda.kanal_ekle(kanal_ismi)
    elif islem == "5":
        print(f"Kanal sayısı: {kumanda.kanal_sayisini_ogrenme()}")
    elif islem == "6":
        kumanda.rastgele_kanal()
    elif islem == "7":
        kumanda.televizyon_bilgileri()
    else:
        print("Geçersiz bir işlem yaptınız, lütfen tekrar deneyin.")
```
